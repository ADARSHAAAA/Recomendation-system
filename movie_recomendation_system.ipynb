```python
df.columns = ['id', 'title', 'cast', 'crew']
df1 = df1.merge(df, on='id')
```

```python
df1['budget'].head(10)
```

```python
c = df1['vote_average'].mean()
c
```

```python
m = df1['vote_count'].quantile(0.9)
m
```

```python
q_movies = df1.copy().loc[df1['vote_count']>=m]
q_movies.shape
```

```python
def weighted_rating(x, m=m, c=c):
    v = x['vote_count']
    R = x['vote_average']

    return (v/(v+m)*R)+(m/(m+v)*c)
```

```python
q_movies['score'] = q_movies.apply(weighted_rating, axis=1)
```

```python
q_movies = q_movies.sort_values('score',ascending=False)
q_movies[['original_title','vote_count','vote_average','score']].head(10)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')
df1['overview'] = df1['overview'].fillna('')
tfidf_matrix = tfidf.fit_transform(df1['overview'])
tfidf_matrix.shape
```

```python
from sklearn.metrics.pairwise import linear_kernel
cosine_similarityy = linear_kernel(tfidf_matrix,tfidf_matrix)
```

```python
indices = pd.Series (df1.index, index=df1['original_title']).drop_duplicates()
```

```python
def get_recommendations(original_title, cosine_similarityy=cosine_similarityy):
  idx = indices[original_title]
  sim_score = list(enumerate(cosine_similarityy[idx]))
  sim_score = sorted(sim_score, key=lambda x:x[1], reverse=True)
  sim_score = sim_score[1:11]
  movie_indices = [i[0] for i in sim_score]
  return df1['original_title'].iloc[movie_indices]

```

```python
get_recommendations('The Dark Knight Rises')
```
